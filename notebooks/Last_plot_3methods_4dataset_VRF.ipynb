```python
import matplotlib.pyplot as plt
import numpy as np

# Define datasets and methods
datasets = ['mnist', 'fmnist', 'cifar10', 'svhn']
methods = ['custom', 'oort']

# Define filenames (old data)
files = {
    'mnist': ('output_mnist_custom_global_dirichlet_Q8_lr0.01_ep1_w0.4.txt', 'output_mnist_oort_global_dirichlet1.txt'),
    'fmnist': ('output_fmnist_custom_global_dirichlet_Q8_lr0.01_ep1_w0.4.txt', 'output_fmnist_oort_global_dirichlet1.txt'),
    'cifar10': ('output_cifar10_custom_global_dirichlet_Q8_lr0.01_ep1_w0.4.txt', 'output_cifar10_oort_global_dirichlet1.txt'),
    'svhn': ('output_svhn_custom_global_dirichlet_Q8_lr0.01_ep1_w0.4.txt', 'output_svhn_oort_global_dirichlet1.txt')
}

base_dir = "/home/kamrul/Documents/kamrul_files_Linux/OORT/ClusterFed_OORT/results/svhn2/"
```

```python
# Function to load metrics from files
def load_metric(file_name, key):
    data = {}
    with open(base_dir + file_name, "r") as f:
        for line in f:
            parts = line.strip().split("=", 1)
            if len(parts) == 2:
                var_name = parts[0].strip()
                try:
                    data[var_name] = eval(parts[1].strip())
                except Exception as e:
                    print(f"Error parsing {var_name} in {file_name}: {e}")
    return data.get(key, [])

# Common font sizes
label_fontsize = 42
legend_fontsize = 34
tick_fontsize = 34


# Colors and styles
colors = {'custom': 'tab:blue', 'oort': 'tab:red'}
linestyles = {'custom': '-', 'oort': '--'}
markers = {'custom': 'o', 'oort': 's'}

# Plotting function
def plot_metric(metric_name, ylabel):
    fig, axes = plt.subplots(2, 2, figsize=(20, 14))
    axes = axes.flatten()

    for idx, dataset in enumerate(datasets):
        ax = axes[idx]

        min_rounds = float('inf')
        metrics = {}

        for method in methods:
            metric_data = load_metric(files[dataset][methods.index(method)], f"{metric_name}_{method}")
            metrics[method] = metric_data
            min_rounds = min(min_rounds, len(metric_data))

        rounds = np.arange(min_rounds)

        for method in methods:
            ax.plot(
                rounds, metrics[method][:min_rounds],
                #label=f"{method.capitalize()}",
                label='Proposed' if method == 'custom' else method.capitalize(),
                color=colors[method],
                linestyle=linestyles[method],
                marker=markers[method],
                markevery=max(1, min_rounds // 10),
                markersize=10,
                linewidth=2
            )

        ax.set_title(dataset.upper(), fontsize=label_fontsize)
        ax.set_xlabel("Rounds", fontsize=label_fontsize)
        ax.set_ylabel(ylabel, fontsize=label_fontsize)
        ax.tick_params(axis='both', which='major', labelsize=tick_fontsize)
        ax.grid(True)

    # Common legend
    # handles, labels = axes[0].get_legend_handles_labels()
    # fig.legend(handles, labels,
    #            loc='upper center', bbox_to_anchor=(0.5, 0.97),
    #            ncol=2, fontsize=legend_fontsize,
    #            frameon=True, fancybox=True, shadow=True)

    # fig.tight_layout(rect=[0, 0.03, 1, 0.94])

    # Individual legends inside each subplot
    for ax in axes:
        ax.legend(loc='lower right', fontsize=legend_fontsize, frameon=True, fancybox=True, shadow=True)

    fig.tight_layout()

    plt.show()

# Plot Accuracy
plot_metric("accuracy", "Accuracy")

# Plot Loss
plot_metric("loss", "Loss")
```
