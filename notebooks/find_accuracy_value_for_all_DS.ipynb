```python
# Define thresholds and display format
thresholds = [60, 65, 70, 75, 80, 85, 90] 
thresholds = [35, 40, 45, 50, 55]

# Function to extract accuracy list
def extract_accuracy_list(file_path, list_name):
    with open(file_path, 'r') as f:
        for line in f:
            if list_name in line:
                return ast.literal_eval(line.split('=')[1].strip())
    raise ValueError(f"{list_name} not found in {file_path}")

# Function to compute round to reach threshold
def round_to_reach_threshold(accuracy_list, threshold):
    for i, acc in enumerate(accuracy_list):
        if acc >= threshold:
            return i + 1
    return float('nan')

# Function to compute improvement
def compute_improvement_ratios(result_table, reference='Proposed'):
    thresholds = [key for key in result_table[reference] if key.startswith('@')]
    baselines = [k for k in result_table if k != reference]

    improvements = {}
    for baseline in baselines:
        time_improvements = []
        for thr in thresholds:
            prop = result_table[reference][thr]
            base = result_table[baseline][thr]
            if not np.isnan(prop) and not np.isnan(base) and prop > 0:
                time_improvements.append(base / prop)

        acc_prop = result_table[reference]['Best Accuracy (%)']
        acc_base = result_table[baseline]['Best Accuracy (%)']
        acc_improvement = acc_prop / acc_base if acc_base > 0 else float('nan')

        improvements[baseline] = {
            'avg_time_to_accuracy_improvement': np.mean(time_improvements) if time_improvements else float('nan'),
            'final_accuracy_improvement': acc_improvement
        }
    return improvements

# Process each dataset
for dataset, paths in files.items():
    print(f"\n=== Dataset: {dataset.upper()} ===")

    result_table = {}
    method_label = {'custom': 'Proposed', 'random': 'Random', 'oort': 'OORT'}

    for method, path in paths.items():
        full_path = base_path + path
        list_name = f"accuracy_{method}"
        accuracy = extract_accuracy_list(full_path, list_name)

        result_table[method_label[method]] = {
            f'@{thr}': round_to_reach_threshold(accuracy, thr) for thr in thresholds
        }
        result_table[method_label[method]]['Best Accuracy (%)'] = max(accuracy)

    # Print result table
    print(f"{'Method':<10} " + " ".join([f"@{t:>3}" for t in thresholds]) + f" {'Best Accuracy (%)':>20}")
    for method, values in result_table.items():
        row = f"{method:<10} " + " ".join([f"{values[f'@{t}']:>5}" for t in thresholds]) + f" {values['Best Accuracy (%)']:>20.2f}"
        print(row)

    # Print improvements
    improvements = compute_improvement_ratios(result_table)
    for method, vals in improvements.items():
        print(f"\nCompared to {method}:")
        if not np.isnan(vals['avg_time_to_accuracy_improvement']):
            print(f"- Time-to-Accuracy Improvement: {vals['avg_time_to_accuracy_improvement']:.2f}X")
        else:
            print("- Time-to-Accuracy Improvement: N/A")
        if not np.isnan(vals['final_accuracy_improvement']):
            print(f"- Final Accuracy Improvement: {vals['final_accuracy_improvement']:.2f}X")
        else:
            print("- Final Accuracy Improvement: N/A")
```
